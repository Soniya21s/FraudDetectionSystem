```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

pd.set_option('display.max_columns',None)##by writting max_columns you can see all columns
```

```python
df = pd.read_csv("../data/raw/transactions.csv")##.. used to comeout of the folder(relative path)
df.head()
```

```python
df.drop(columns="transaction id",inplace= True)
```

```python
df.head(2)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], dayfirst=True)
```

```python
df.head(2)
```

```python
df['is_night'] = df['hour_of_day'].between(0, 5).astype(int)
```

```python
df.head()
```

```python
df[df["is_night"] == 1]
```

```python
df['is_office_hours'] = df['hour_of_day'].between(9, 18).astype(int)
```

```python
df.head()
```

```python
df['log_amount'] = np.log1p(df['amount (INR)'])
```

```python
high_amt_threshold = df['amount (INR)'].quantile(0.99)

df['is_high_amount'] = (df['amount (INR)'] >= high_amt_threshold).astype(int)
```

```python
df.head()
```

```python
df[df["is_high_amount"] == 1]
```

```python
df['same_bank'] = (df['sender_bank'] == df['receiver_bank']).astype(int)
```

```python
df['bank_pair'] = df['sender_bank'] + "_" + df['receiver_bank']
```

```python
df.head()
```

```python
df['device_risk'] = df['device_type'].map({
    'Web': 'High',
    'Android': 'Medium',
    'iOS': 'Low'
})
```

```python
df['network_risk'] = df['network_type'].map({
    'WiFi': 'High',
    '3G': 'Medium',
    '4G': 'Low',
    '5G': 'Low'
})
```

```python
df['device_network_risk'] = (
    df['device_risk'] + "_" + df['network_risk']
)
```

```python
df.head()
```

```python
age_map = {
    '18-25': 1,
    '26-35': 2,
    '36-45': 3,
    '46-55': 4,
    '56+': 5
}

df['sender_age_ord'] = df['sender_age_group'].map(age_map)
df['receiver_age_ord'] = df['receiver_age_group'].map(age_map)
```

```python
df['age_gap'] = abs(
    df['sender_age_ord'] - df['receiver_age_ord']
)
```

```python
state_freq = df['sender_state'].value_counts(normalize=True)

df['sender_state_freq'] = df['sender_state'].map(state_freq)
```

```python
df.head()
```

```python
df[['is_night', 'is_high_amount', 'same_bank']].value_counts()
```

```python
df.select_dtypes(include='number').corr()
```

```python
X = df.drop(columns=['fraud_flag'])
y = df['fraud_flag']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
def frequency_encode(train_col, test_col):
    freq = train_col.value_counts(normalize=True)
    return train_col.map(freq), test_col.map(freq).fillna(0)
```

```python
freq_cols = [
    'merchant_category',
    'sender_bank',
    'receiver_bank',
    'bank_pair',
    'sender_state'
]

for col in freq_cols:
    X_train[col], X_test[col] = frequency_encode(
        X_train[col], X_test[col]
    )
```

```python
X_train
```

```python
ohe_cols = [
    'transaction type',
    'transaction_status',
    'device_type',
    'network_type',
    'day_of_week',
    'device_risk',
    'network_risk',
    'device_network_risk'
]
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

ohe.fit(X_train[ohe_cols])
```

```python
# Diagnostic: check which ohe columns are missing
missing = [c for c in ohe_cols if c not in X_train.columns]
print('Missing OHE columns:', missing)
print('\nX_train columns sample:', list(X_train.columns)[:40])
```

```python
X_train_ohe = ohe.transform(X_train[ohe_cols])
X_test_ohe = ohe.transform(X_test[ohe_cols])
```

```python
X_train_num = X_train.drop(columns=ohe_cols)
X_test_num = X_test.drop(columns=ohe_cols)
```

```python
# Remove sender/receiver age groups and any remaining categorical columns
# (object or category dtypes), then update numeric and final feature sets
cols_to_remove = ['sender_age_group', 'receiver_age_group']
# detect remaining categorical columns
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
# include ohe columns too (they will be replaced by ohe output)
cols_to_remove = list(dict.fromkeys(ohe_cols + cols_to_remove + cat_cols))
cols_present = [c for c in cols_to_remove if c in X_train.columns]
print('Dropping columns:', cols_present)

# Update numeric datasets
X_train_num = X_train.drop(columns=cols_present)
X_test_num = X_test.drop(columns=cols_present)
print('X_train_num shape:', X_train_num.shape)
print('X_test_num shape:', X_test_num.shape)

# Ensure no categorical columns remain in the numeric sets
print('Remaining object/category columns in X_train_num:',
      X_train_num.select_dtypes(include=['object', 'category']).columns.tolist())

# Convert OHE output to DataFrame and concatenate to numeric features to form final datasets
ohe_feature_names = ohe.get_feature_names_out(ohe_cols)
X_train_ohe_df = pd.DataFrame(X_train_ohe, columns=ohe_feature_names, index=X_train.index)
X_test_ohe_df = pd.DataFrame(X_test_ohe, columns=ohe_feature_names, index=X_test.index)

X_train_final = pd.concat([X_train_num, X_train_ohe_df], axis=1)
X_test_final = pd.concat([X_test_num, X_test_ohe_df], axis=1)
print('Final feature shapes -> train:', X_train_final.shape, ', test:', X_test_final.shape)
```
